# Beat UCB and Thompson Sampling

In this notebook you will play against two classic bandit strategies:
- UCB
- Thompson sampling

The reward on each pull is always an integer in `{0, 1, ..., 10}`.
The hidden reward distributions are created randomly when the game starts.

Your goal is simple: finish with a higher cumulative reward than both algorithms.


## Rules

Each round has three players:
- you
- UCB
- Thompson sampling

At the start of each round, the notebook shows which arm UCB would choose and which arm Thompson sampling would choose.
Then you decide what to do.

You can:
- choose an arm directly, such as `0`, `1`, `2`, ...
- set `user_choice = "ucb"` to follow UCB
- set `user_choice = "thompson"` to follow Thompson sampling

The notebook keeps score with cumulative rewards only. It does not reveal regret or the hidden reward distributions.
