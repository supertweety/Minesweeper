# Minesweeper Tournament Dashboard

This notebook demonstrates the turn-based Minesweeper tournament framework for symbolic or SAT-based agents.

It includes:

- a sample tournament run
- an interactive round-and-turn viewer for the board state
- per-turn game progress information
- cumulative agent statistics across the whole tournament

The dashboard is built with plain notebook HTML and JavaScript, so you can inspect what happened in each round and debug your own agents.

````python
from functools import partial

from IPython.display import Markdown, display

from minesweeper_tournament import (
    CrashAgent,
    LocalInferenceAgent,
    RandomAgent,
    RowMajorAgent,
    SlowAgent,
    TournamentConfig,
    TournamentRunner,
)
from minesweeper_tournament.dashboard import (
    build_dashboard_payload,
    render_tournament_dashboard,
)
````

## Run a Sample Tournament

The next cell runs a small tournament with a few baseline agents. You can later swap these out for your own SAT-based agents while keeping the same visualization cell.

````python
runner = TournamentRunner(
    agent_factories={
        'local': partial(LocalInferenceAgent, seed=7, name='local'),
        'row_major': partial(RowMajorAgent, name='row_major'),
        'random_a': partial(RandomAgent, seed=1, name='random_a'),
        'random_b': partial(RandomAgent, seed=2, name='random_b'),
    },
    config=TournamentConfig(
        board_size=5,
        num_mines=5,
        num_rounds=100,
        turn_timeout_seconds=0.3,
        random_seed=123,
    ),
)

result = runner.run()

display(Markdown(
    f"**Rounds played:** {len(result.rounds)}  \n"
    f"**Final scores:** `{result.scores}`  \n"
    f"**Disqualified agents:** `{result.disqualified_agents}`"
))
````

## Interactive Dashboard

Use the round selector and turn slider to scrub through the tournament. The dashboard shows:

- the visible board after the selected turn
- the current event and score table
- a round event log with the current step highlighted
- cumulative score lines across the full tournament
- cumulative per-agent counts for safe moves, mine hits, timeouts, crashes, and invalid moves


````python
payload = build_dashboard_payload(result)
render_tournament_dashboard(payload)
````

## Swap in Your Own Agents

Keep the visualization cell exactly as it is and only replace the `agent_factories` dictionary.

For example, if you implement a SAT-based agent class `SatAgent`, run:

```python
runner = TournamentRunner(
    agent_factories={
        'sat_agent': SatAgent,
        'local': partial(LocalInferenceAgent, seed=7, name='local'),
        'row_major': partial(RowMajorAgent, name='row_major'),
        'random_a': partial(RandomAgent, seed=1, name='random_a'),
    },
    config=TournamentConfig(
        board_size=5,
        num_mines=5,
        num_rounds=20,
        turn_timeout_seconds=0.5,
        random_seed=123,
    ),
)
result = runner.run()
payload = build_dashboard_payload(result)
render_tournament_dashboard(payload)
```
